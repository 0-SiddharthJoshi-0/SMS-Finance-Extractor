```python
import pandas as pd

# Load your dataset
df = pd.read_csv("TransactionMemo.csv")

# Select 400 messages randomly
sample_df = df[['SMS']].sample(n=400, random_state=42).reset_index(drop=True)

# Rename the column to 'text' (Label Studio expects this)
sample_df.rename(columns={"SMS": "text"}, inplace=True)

# Save to CSV
#sample_df.to_csv("Labeled_SMS.csv", index=False)
```

```python
import ast

# Keep only 'text' and 'label' columns
annotated_df = annotated_df[['text', 'label']].copy()

# Convert label from string to list-of-dicts (if needed)
annotated_df['label'] = annotated_df['label'].apply(
    lambda x: ast.literal_eval(x) if isinstance(x, str) else x
)
```

```python
import os
import torch
from transformers import BertTokenizerFast

# Suppress warnings
os.environ["TOKENIZERS_PARALLELISM"] = "false"
os.environ["WANDB_DISABLED"] = "true"

# Label Normalization
LABEL_NORMALIZATION = {
    "AMOUNT": "Amount",
    "CURRENCY": "Currency",
    "CREDIT_OR_DEBIT": "Credit_or_Debit",
    "ACCOUNT_NUMBER": "AccoutNo",
    "BALANCE": "Balance",
    "LOCATION": "Location"
}

# Label Mappings
all_labels = ['O']
for label in LABEL_NORMALIZATION.values():
    all_labels.extend([f'B-{label}', f'I-{label}'])

label2id = {label: i for i, label in enumerate(sorted(set(all_labels)))}
id2label = {i: label for label, i in label2id.items()}

# Load tokenizer
tokenizer = BertTokenizerFast.from_pretrained("bert-base-uncased")
```

```python
def char_to_bio(text, entities):
    tags = ['O'] * len(text)
    for ent in entities:
        start, end = ent['start'], ent['end']
        raw_label = ent['labels'][0]
        label = LABEL_NORMALIZATION.get(raw_label.upper())
        if label is None: continue
        tags[start] = f'B-{label}'
        for i in range(start+1, end):
            tags[i] = f'I-{label}'
    return tags

def tokenize_and_align_labels(row):
    text = row['text']
    entities = row['label']
    char_tags = char_to_bio(text, entities)
    encoding = tokenizer(text, return_offsets_mapping=True, truncation=True, max_length=128)
    labels = []
    for idx, (start, end) in enumerate(encoding['offset_mapping']):
        if start == end:
            labels.append(label2id['O'])
        else:
            tag = char_tags[start] if start < len(char_tags) else 'O'
            labels.append(label2id.get(tag, label2id['O']))
    encoding['labels'] = labels
    return encoding

tokenized_dataset = [tokenize_and_align_labels(row) for _, row in annotated_df.iterrows()]
```

```python
from datasets import Dataset
import numpy as np

dataset = Dataset.from_dict({
    'input_ids': [d['input_ids'] for d in tokenized_dataset],
    'attention_mask': [d['attention_mask'] for d in tokenized_dataset],
    'labels': [d['labels'] for d in tokenized_dataset]
})

# 80/20 train/test split
dataset = dataset.train_test_split(test_size=0.2, seed=42)
```

```python
from transformers import BertForTokenClassification, TrainingArguments, Trainer, DataCollatorForTokenClassification

model = BertForTokenClassification.from_pretrained(
    "bert-base-uncased",
    num_labels=len(label2id),
    id2label=id2label,
    label2id=label2id
)

data_collator = DataCollatorForTokenClassification(tokenizer)

training_args = TrainingArguments(
    output_dir="./bert-sms-ner-2",
    do_train=True,
    do_eval=True,
    learning_rate=2e-5,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    num_train_epochs=3,
    weight_decay=0.01
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=dataset['train'],
    eval_dataset=dataset['test'],
    data_collator=data_collator
)

trainer.train()
```

```python
from seqeval.metrics import classification_report
import numpy as np

def align_predictions(predictions, label_ids):
    preds = np.argmax(predictions, axis=2)
    batch_size, seq_len = preds.shape
    out_label_list = [[] for _ in range(batch_size)]
    preds_list = [[] for _ in range(batch_size)]
    for i in range(batch_size):
        for j in range(seq_len):
            if label_ids[i][j] != -100:
                out_label_list[i].append(id2label[label_ids[i][j]])
                preds_list[i].append(id2label[preds[i][j]])
    return preds_list, out_label_list

outputs = trainer.predict(dataset['test'])
preds_list, out_label_list = align_predictions(outputs.predictions, outputs.label_ids)
print(classification_report(out_label_list, preds_list))
```

```python
def predict_entities(text, model, tokenizer, label_map):
    model.eval()
    encoding = tokenizer(text, return_tensors="pt", truncation=True, max_length=128, return_offsets_mapping=True)
    offset_mapping = encoding.pop("offset_mapping")

    # Move inputs to the same device as the model
    encoding = {k: v.to(model.device) for k, v in encoding.items()}

    with torch.no_grad():
        outputs = model(**encoding)

    predictions = torch.argmax(outputs.logits, dim=2)[0].cpu().tolist()  # .cpu() to move back to CPU
    offsets = offset_mapping[0].tolist()

    result = []
    for idx, (start, end) in enumerate(offsets):
        if start == end:
            continue
        label_id = predictions[idx]
        label = label_map[label_id]
        if label != 'O':
            word = text[start:end]
            result.append((word, label))
    return result
```

```python
# Take first 5 SMS messages from the DataFrame
sms_samples = df['SMS'].head(5).tolist()

# Predict entities for each SMS
for i, sms in enumerate(sms_samples, 1):
    print(f"\nSMS {i}: {sms}")
    predicted_entities = predict_entities(sms, model, tokenizer, id2label)
    
    if predicted_entities:
        print("Predicted Entities:")
        for word, label in predicted_entities:
            print(f"  {word}: {label}")
    else:
        print("No entities predicted.")
```

```python
sample_text = "INR 500.00 was debited from your account *XXXX1234. Available balance: INR 3050.00"

predicted_entities = predict_entities(sample_text, model, tokenizer, id2label)

print("Predicted Entities:")
for word, label in predicted_entities:
    print(f"{word}: {label}")
```
